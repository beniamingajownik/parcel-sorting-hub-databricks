# Create Bronze Schema

```sql
CREATE SCHEMA IF NOT EXISTS parcel_sorting_hub.bronze;
```

# Create Tables using CTAS (Run once)

```sql
CREATE OR REPLACE TABLE parcel_sorting_hub.bronze.devices 
USING DELTA 
AS

SELECT 
    *,
    _metadata.file_name AS _source_file,
    current_timestamp() AS _ingested_at
FROM read_files(
    's3://sortownia-data/CTAS-raw-data/devices.csv',
    format => 'CSV',
    header => TRUE,
    inferSchema => FALSE
);

CREATE OR REPLACE TABLE parcel_sorting_hub.bronze.exceptions 
USING DELTA 
AS

SELECT 
    *,
    _metadata.file_name AS _source_file,
    current_timestamp() AS _ingested_at
FROM read_files(
    's3://sortownia-data/CTAS-raw-data/2026-05-01_exceptions.csv',
    format => 'CSV',
    header => TRUE,
    inferSchema => FALSE
);

CREATE OR REPLACE TABLE parcel_sorting_hub.bronze.intake 
USING DELTA 
AS

SELECT 
    *,
    _metadata.file_name AS _source_file,
    current_timestamp() AS _ingested_at
FROM read_files(
    's3://sortownia-data/CTAS-raw-data/2026-05-01_intake.csv',
    format => 'CSV',
    header => TRUE,
    inferSchema => FALSE
);

CREATE OR REPLACE TABLE parcel_sorting_hub.bronze.loading 
USING DELTA 
AS

SELECT 
    *,
    _metadata.file_name AS _source_file,
    current_timestamp() AS _ingested_at
FROM read_files(
    's3://sortownia-data/CTAS-raw-data/2026-05-01_loading.csv',
    format => 'CSV',
    header => TRUE,
    inferSchema => FALSE
);

CREATE OR REPLACE TABLE parcel_sorting_hub.bronze.parcels 
USING DELTA 
AS

SELECT 
    *,
    _metadata.file_name AS _source_file,
    current_timestamp() AS _ingested_at
FROM read_files(
    's3://sortownia-data/CTAS-raw-data/2026-05-01_parcels.csv',
    format => 'CSV',
    header => TRUE,
    inferSchema => FALSE
);

CREATE OR REPLACE TABLE parcel_sorting_hub.bronze.scans 
USING DELTA 
AS

SELECT 
    *,
    _metadata.file_name AS _source_file,
    current_timestamp() AS _ingested_at
FROM read_files(
    's3://sortownia-data/CTAS-raw-data/2026-05-01_scans.csv',
    format => 'CSV',
    header => TRUE,
    inferSchema => FALSE
);

CREATE OR REPLACE TABLE parcel_sorting_hub.bronze.sorting 
USING DELTA 
AS

SELECT 
    *,
    _metadata.file_name AS _source_file,
    current_timestamp() AS _ingested_at
FROM read_files(
    's3://sortownia-data/CTAS-raw-data/2026-05-01_sorting.csv',
    format => 'CSV',
    header => TRUE,
    inferSchema => FALSE
);

CREATE OR REPLACE TABLE parcel_sorting_hub.bronze.status_history 
USING DELTA 
AS

SELECT 
    *,
    _metadata.file_name AS _source_file,
    current_timestamp() AS _ingested_at
FROM read_files(
    's3://sortownia-data/CTAS-raw-data/2026-05-01_status_history.csv',
    format => 'CSV',
    header => TRUE,
    inferSchema => FALSE
);
```
